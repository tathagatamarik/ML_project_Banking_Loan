```python
from sklearn.pipeline import Pipeline
##feature Scaling
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
## Pieplining
numeric_preprocessor = Pipeline(
    steps=[
        ("imputation_mean", SimpleImputer(missing_values=np.nan, strategy="mean")),
        ("scaler", StandardScaler()),
    ]
)
```

```python
from sklearn import set_config
```

```python
set_config(display='diagram')
```

```python
numeric_preprocessor
```

```python
categorical_preprocessor = Pipeline(
    steps=[
        (
            "imputation_constant",
            SimpleImputer(fill_value="missing", strategy="constant"),
        ),
        ("onehot", OneHotEncoder(handle_unknown="ignore")),
    ]
)
```

```python
categorical_preprocessor
```

```python
preprocessor=ColumnTransformer(
    [("categorical",categorical_preprocessor,categorical_columns),
    ("numerical",numeric_preprocessor,numerical_columns)])
```

```python
preprocessor
```

## Hyper parameter tuning

```python
new_lr_pipe = Pipeline([
    ('pre',preprocessor ),
    ('lr', LogisticRegression())
])
```

```python
param_grids={'lr__C':[0.001,0.1,2,10],
             'lr__penalty':['l1','l2']}
```

```python
gs=GridSearchCV(new_lr_pipe,param_grids)
gs.fit(X_train,y_train)
```

```python
test_score = gs.score(X_test,y_test)
print("test score:",test_score)
print("best parameters: ",gs.best_params_)
print("best score: ", gs.best_score_)
```

```python
import xgboost as xgb
```

```python
new_xgboost_pipe = Pipeline([
    ('pre',preprocessor ),
    ('xgb', xgb.XGBClassifier())
])
```

```python
param_grid = {'xgb__learning_rate': [0.1, 0.01, 0.001],
              'xgb__max_depth': [3, 5, 7]}

grid = GridSearchCV(new_xgboost_pipe, param_grid, cv=5)
grid.fit(X_train, y_train)
```

```python
test_score1 = grid.score(X_test,y_test)
print("test score:",test_score1)
print("best parameters: ",grid.best_params_)
print("best score: ", grid.best_score_)
```

```python
new_rf_pipe = Pipeline([
    ('pre',preprocessor ),
    ('rf', RandomForestClassifier(n_estimators = 10,
                             criterion = 'entropy',
                             random_state =7))
])
```
